Here, fake and real dataframes are combined,at first fake news dataframe with 16817 rows,then real dataframe's 5323 rows until the end.

```python
stop_words = set(stopwords.words('english'))
```

```python
def preprocess_text(text):  # Preprocessing setup
  text = str(text).lower() # lowercase
  text = re.sub(f"[{re.escape(string.punctuation)}]", "", text) # remove punctuation
  tokens = text.split() # tokenize

  # Remove stopwords and short tokens
  cleaned_tokens = [t for t in tokens if t not in stop_words and len(t) > 2]
  return " ".join(cleaned_tokens)


print("Cleaning started , please wait...")
df["clean_title"] = df["title"].apply(preprocess_text) # creating new column with cleaned titles using our function

print("Before Cleaning / After Cleaning")
pd.set_option('display.max_colwidth', 100)
display(df[['title', 'clean_title', 'label']].head(10))
```

Now we have fake and real titles in order. In order to do train test split part as we want, we need to mix these data. And we need to delete repeated news, because we do not want our model to memorise patterns.

```python
df['clean_title'].replace(r'^\s*$', np.nan, regex=True, inplace=True) # replacing empty rows with NaN , and using regex (regular expression)
```

```python
null_count = df['clean_title'].isnull().sum()
print(f"Number of empty rows after cleaning: {null_count}")
```

Here, we have 4 empty rows after second cleaning.

```python
if null_count > 0:
    df.dropna(subset=['clean_title'], inplace=True)
    print("Null Rows (empty titles) are dropped with dropna")
```

```python
# to eliminate empty rows

empty_rows = df[df['clean_title'].str.strip() == ""].index
print(f"Number of empty Rows: {len(empty_rows)}")
```

```python
# to mix our dataframe
df = df.sample(frac=1, random_state=42).reset_index(drop=True)
print("Dataframe is mixed up.")
```

Now we have fake and real titles in order. In order to do train test split part as we want, we need to mix these data. And we need to delete repeated news, because we do not want our model to memorise patterns.

```python
duplicate_count = df.duplicated(subset=['clean_title']).sum()
print(f"Number of duplicate rows: {duplicate_count}")
```

```python
if duplicate_count > 0:
    df.drop_duplicates(subset=['clean_title'], keep='first', inplace=True)
    print("Repeated news are dropped.")
```

```python
df = df.sample(frac=1, random_state=42).reset_index(drop=True)
print("Dataframe is shuffled and index reset.")
```

```python
df.to_csv("cleaned_news.csv", index=False)
print("Final CSV saved: cleaned_news.csv")
```
